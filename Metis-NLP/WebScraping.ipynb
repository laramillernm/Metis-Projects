```python
# scrape nm bills with governor action from 2013 to 2020
# unidecode imported 
##with open('nmbills_13_20v3.csv','w', encoding='utf-8') as file: 
    ##file.write("Bill_ID, Title, Sponsor, Last_Action, Session \n")

driver = webdriver.Chrome(r"C:\Users\olive\Documents\Metis\ASLearning\webscraping\chromedriver_win32\chromedriver.exe")

driver.get('https://www.nmlegis.gov/Legislation/BillFinder/Governor_Actions')
driver.maximize_window()
time.sleep(1)

dropdown_f = driver.find_element_by_xpath('//*[@id="MainContent_ddlSessionStart"]/option[18]')
dropdown_f.click()

dropdown_t = driver.find_element_by_xpath('//*[@id="MainContent_ddlSessionEnd"]/option[6]')
dropdown_t.click()

go = driver.find_element_by_xpath('//*[@id="MainContent_btnSearch"]')
go.click()
time.sleep(2)

### download whole table
### code written by SIM and can be found on stackoverflow
### https://stackoverflow.com/questions/52448137/python-selenium-scrape-the-whole-table
item = driver.find_element_by_xpath('//*[@id="MainContent_gridViewLegislation"]')
driver.execute_script("arguments[0].click();", item)
time.sleep(2)
df = pd.DataFrame(columns = ['Bill ID', 'Title', 'Sponsor', 'Last Action', 'Session'] )
BillList = []
for table in driver.find_elements_by_xpath('//*[@id="MainContent_gridViewLegislation"]//tr'):
    data = [item.text for item in table.find_elements_by_xpath(".//*[self::td or self::th]")]
    #df.append(data)
    BillList.append(data)
    #print(data)

driver.quit()
 
```

```python
df_BillList = pd.DataFrame(BillList[1:], columns = BillList[0])
df_BillList.head(20)
```

```python
df_BillList['Year']= (df_BillList['Session'].str[:4]).astype(int)
df_BillList.head()
```

```python
df_BillList['Bill ID'] = df_BillList['Bill ID'].str.replace('*', '')
df_BillList['Bill ID'].unique()
```

```python
df_BillList['Last Action'].unique()
```

```python
df_BillList['SessionType'] = df_BillList['Session'].str[5:]
df_BillList['SessionType'].unique()
```

```python
df_BillList['SessionType'] = df_BillList['SessionType'].str.replace('1st Special', 'Special')
df_BillList['SessionType'] = df_BillList['SessionType'].str.replace('2nd Special', 'Special2')

df_BillList.head(11)
```

```python
df_BillList['BillNum']=df_BillList['Bill ID'].str[3:]
df_BillList['BillNum'] = df_BillList['BillNum'].apply(lambda x: x.zfill(4))
df_BillList.sample()
```

```python
df_BillList['Bill_Link'] = ('https://www.nmles.gov/Sessions/' +df_BillList['Session'].str[2:4] +'%20' + df_BillList['SessionType']+'/final/'+ 
                    df_BillList['Bill ID'].str[:2])+df_BillList['BillNum']+'.pdf'
### example link https://www.nmlegis.gov/Sessions/19%20Regular/final/HB0005.pdf
pd.set_option('display.max_colwidth', 500)
df_BillList.head()
```

```python
df_BillList['filename']='Yr_'+df_BillList.Bill_Link.str[34:36]+df_BillList.Bill_Link.str[36:37]+df_BillList.Bill_Link.str.split('/').str.get(-1)
df_BillList.sample()
```

```python
df_BillList.to_csv('BillText.csv')###save changes to scraped table data
```

```python
link_list = df_BillList['Bill_Link'].to_list()
link_list
```

```python
for link in link_list: 
  
        '''iterate through all links in in list 
        and download them one by one'''
          
        # obtain filename by adding year and abbr for session type
        # splitting url and getting last string 
        file_name = 'Yr_'+link[33:35]+link[38:39]+link.split('/')[-1] 
  
        print( "Downloading file:%s"%file_name)  ## this is check so I go go to links that aren't working and troubleshoot
          
        # create response object 
        r = requests.get(link) 
          
        # download started 
        with open(file_name, 'wb') as f:
        #with open(file_name, 'wb') as f: 
            for chunk in r.iter_content(decode_unicode=True): 
                try:
                    if chunk:
                        f.write(chunk)
                except:
                    continue
          
        print( "%s downloaded!\n"%file_name ) ##download check
print ("All bills downloaded")
```
